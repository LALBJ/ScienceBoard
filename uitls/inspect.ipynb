```python
from PIL import Image, ImageDraw

def mark_points_on_image(image_path, points):
    """
    在图片上根据坐标点标注圆点并保存新图片。

    :param image_path: 图片路径
    :param points: 坐标列表 [(x1, y1), (x2, y2), ...]
    :param output_path: 标注后图片保存路径
    """
    # 打开图片
    image = Image.open(image_path)
    draw = ImageDraw.Draw(image)

    # 标注圆点参数
    radius = 5
    fill_color = (255, 0, 0)  # 红色

    for (x, y) in points:
        left_up_point = (x - radius, y - radius)
        right_down_point = (x + radius, y + radius)
        draw.ellipse([left_up_point, right_down_point], fill=fill_color)

    # 保存图片
    return image

```

```python
image_path = "/data/code/ScienceBoard/logs/tars_dpo-vm-screenshot/TeXstudio/C-04/step_1@250626105612.png"
# 示例坐标，可以替换成你的输入
points = [(742,376)]  

mark_points_on_image(image_path, points)
```
